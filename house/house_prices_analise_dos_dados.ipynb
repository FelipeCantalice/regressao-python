### Extrair os dados de sqft_living

```python
# sqft_living
X_cases = data.iloc[:,5:6].values
X_cases
```

### Extrair os dados de price

```python
# price
y_casas = data.iloc[:,2].values
y_casas
```

### Divis√£o base de treinamento e teste

```python
from sklearn.model_selection import train_test_split
# 30%
X_casa_treinamento, X_casa_teste, y_casa_treinamento, y_casa_teste = train_test_split(X_cases, y_casas, test_size=0.3, random_state=0)
```

```python
X_casa_treinamento.shape, y_casa_treinamento.shape
```

```python
X_casa_teste.shape, y_casa_teste.shape
```

### Aplica√ß√£o da regress√£o linear

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_casa_treinamento, y_casa_treinamento)
```

### Resultado

```python
# y = a + bx
print(f'a = {model.intercept_}, b = {model.coef_}')
```

### Variabilidade dos dados

```python
score = model.score(X_casa_treinamento, y_casa_treinamento)
score
```

```python
score = model.score(X_casa_teste, y_casa_teste)
score
```

### Previs√µes

```python
previsoes = model.predict(X_casa_treinamento)
previsoes
```

### Linha de tendencia

```python
figure = px.scatter(x=X_casa_treinamento.ravel(),y=previsoes)
figure.show()
```

### Diagrama de dispers√£o

```python
figure1 = px.scatter(x=X_casa_treinamento.ravel(),y=y_casa_treinamento)

figure2 = px.line(x=X_casa_treinamento.ravel(),y=previsoes)
figure2.data[0].line.color = 'red'

figure3 = go.Figure(data=figure1.data + figure2.data)
figure3
```

### Metricas

```python
previsoes_teste = model.predict(X_casa_teste)
previsoes_teste
```

```python
y_casa_teste
```

## Erro

Vemos abaixo que o algoritmo pode errar R$: 172985 para mais ou para menos. Isso √© bem ruim.

```python
abs(y_casa_teste - previsoes_teste).mean()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mae = mean_absolute_error(y_casa_teste, previsoes_teste)
mse = mean_squared_error(y_casa_teste, previsoes_teste)

print(f'mae: {mae}')
print(f'mse: {mse}')
```

```python
figure1 = px.scatter(x=X_casa_teste.ravel(),y=y_casa_teste)

figure2 = px.line(x=X_casa_teste.ravel(),y=previsoes_teste)
figure2.data[0].line.color = 'red'

figure3 = go.Figure(data=figure1.data + figure2.data)
figure3
```
